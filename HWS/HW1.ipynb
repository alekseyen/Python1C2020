```python
def create_dict(texts, dictionary = None):
    dictionary = dictionary or {}
    for text_line in texts:
        for word in text_line.split(' '):
            if word not in dictionary:
                dictionary[word] = len(dictionary)
    return dictionary
            
text = [", "]
text2 = [" , ", "   "]

d1 = create_dict(text)
d2 = create_dict(text2, d1)

print(d2, d1 is d2)
```
